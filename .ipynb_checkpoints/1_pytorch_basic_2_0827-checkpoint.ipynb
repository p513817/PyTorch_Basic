# 使用nn.Linear進行修改

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

import torch
import numpy as np
import matplotlib.pyplot as plt

''' 設定好 輸入、權重、輸出 '''
n = 100
d = 1
x = torch.randn(n,d)
w = torch.tensor([10.])         # 固定 w 為 ground truth
b = torch.tensor([3.])          # 固定 b 為 ground truth
y = w * x + b + torch.randn(x.shape)*0.5             # 簡單的線性方程，取得 10 個點的 y 座標 再加上一點雜訊

''' 畫圖 '''
plt.figure(dpi=100)    # 宣告一個圖片 dpi 設為 100
plt.scatter(x.detach().numpy(), y.detach().numpy(), color= 'r')    # 將正確答案先畫出來標示為紅色
attributes = ['--k','--b','--c','--m','--y','--g', '--r']    # 顏色線條參數

''' 初始化參數 '''
x_ = torch.randn(x.shape)         # 宣告一組新的 x_ 等等來拿預測用
model = nn.Linear(d, d)                    # 利用 torch.nn 宣告一個線性模型
epochs = 50                                # 總共跑 N 次迭代
log_time = epochs/5                        # 總共會記錄 5 次 loss 
lr = 0.1                                   # lr 是學習率

for i in range(epochs + 1):
    
    y_ = model(x)
    
    loss_fn = nn.MSELoss()             # 實例化 MSE
    
    loss = loss_fn(y, y_)            # 計算 loss
    
    loss.backward()               # 進行反向傳遞，獲得梯度
    
    with torch.no_grad():
        
        model.weight -= lr * model.weight.grad        # 透過梯度、學習率來更新權重 
        model.bias -= lr * model.bias.grad
        
        model.weight.grad.zero_()           # 記得要將梯度歸零否則會累計
        model.bias.grad.zero_()
        
        if(i%log_time==0):                # 每 5 次 印出 loss 跟 w_ 的狀況
            
            print('[{:>02}/{}] loss:{:5.3f} \t w_:{} \t b_:{}'.format(i,epochs,
                                                               loss.data,
                                                               model.weight.data,
                                                               model.bias.data))    # {:5.3f} 總寬度 5 小數點後 3 位
            
            plt.plot(x_.detach().numpy(), model(x_).detach().numpy() , attributes[int(i/log_time)], label=str(i))    # 畫出曲線

plt.legend(loc='best')    # 自動調整最佳位置
plt.show()
```

# 使用 nn.Module以及optim 對於 linear regression進行改造


```python
import torch
from torch import nn
from torch import optim
import matplotlib.pyplot as plt
from torch import optim

class Model(nn.Module):
    
    def __init__(self, input_dim, out_dim):
        super().__init__()
        self.lin = nn.Linear(input_dim, out_dim)

    def forward(self, x):
        return self.lin(x)

''' 設定好 輸入、權重、輸出 '''
n, d = 100, 1
x = torch.randn(n,d)
w = torch.tensor([10.])         # 固定 w 為 ground truth
b = torch.tensor([3.])          # 固定 b 為 ground truth
y = w * x + b + torch.randn(x.shape)*0.5  

''' 初始化參數 '''
x_ = torch.randn(x.shape)         # 宣告一組新的 x_ 等等來拿預測用
epochs = 50                       # 總共跑 N 次迭代
log_time = epochs/5               # 總共會記錄 5 次 loss 
lr = 0.1                          # lr 是學習率

''' 設定模型以及優化器'''
model = Model(d, d)                              # 實例化 nn.Module
opt = optim.SGD(model.parameters(), lr = lr)     # 使用 SGD 優化器

''' 訓練 '''
for i in range(epochs +1):
    
    y_ = model(x)
    loss_fn = nn.MSELoss()        # 實例化 MSE
    
    loss = loss_fn(y, y_)         # 計算 loss
    loss.backward()               # 進行反向傳遞，獲得梯度
    
    opt.step()                    # 透過優化器去更新參數
    opt.zero_grad()               # 將梯度歸零
        
    if(i% log_time == 0 ):      
        
        # 提取資料都加上 detach()，確保資料不會被重複計算
        print('[{:3}/{:3}] \t loss:{:6.3f}'.format(i, epochs, loss.detach()))   
```

# 動畫

```python
import torch
from torch import nn
from torch import optim
import matplotlib.pyplot as plt
from torch import optim
import numpy as np

import matplotlib.animation as animation

class Model(nn.Module):
    
    def __init__(self, input_dim, out_dim):
        super().__init__()
        self.lin = nn.Linear(input_dim, out_dim)

    def forward(self, x):
        return self.lin(x)

''' 畫圖 '''
fig = plt.figure()
plt.scatter(x, y, color= 'r')    # 將正確答案先畫出來標示為紅色
attributes = ['--k','--b','--c','--m','--y','--g', '--r']    # 顏色線條參數
imgs = []

''' 設定好 輸入、權重、輸出 '''
n, d = 100, 1
x = torch.randn(n,d)
w = torch.tensor([10.])         # 固定 w 為 ground truth
b = torch.tensor([3.])          # 固定 b 為 ground truth
y = w * x + b + torch.randn(x.shape)*0.5  

''' 初始化參數 '''
x_ = torch.randn(x.shape)         # 宣告一組新的 x_ 等等來拿預測用
epochs = 50                     # 總共跑 N 次迭代
log_time = epochs/5               # 總共會記錄 5 次 loss 
lr = 0.1                          # lr 是學習率

''' 設定模型以及優化器'''
model = Model(d, d)
SGD = optim.SGD(model.parameters(), lr = lr)     # 使用 SGD 優化器
Ada = optim.Adagrad(model.parameters(), lr = lr) # 使用 Adagrad 優化器
RMS = optim.RMSprop(model.parameters(), lr = lr)  # 使用 RMSprop 優化器
Adam = optim.Adam(model.parameters(), lr = lr)     # 使用 SGD 優化器
opt = SGD

''' 訓練 '''
for i in range(epochs + 1):
    
    y_ = model(x)
    
    loss_fn = nn.MSELoss()        # 實例化 MSE
    
    loss = loss_fn(y, y_)         # 計算 loss
    
    loss.backward()               # 進行反向傳遞，獲得梯度
    
    opt.step()                    # 透過優化器去更新參數
    
    opt.zero_grad()               # 將梯度歸零
    
    ''' 畫圖 '''
    line = plt.plot(x_.detach().numpy(), model(x_).detach().numpy() , attributes[int(i/log_time)])    # 畫出曲線
    imgs.append(line)
    
ani = animation.ArtistAnimation(fig, imgs, interval=10, blit=True,
                                repeat_delay=1000)    
    
ani.save("linear_regression_{}.gif".format('SGD') , writer='pillow')
```

# Activation Function



```python
import torch
from torch import nn

sigmoid = nn.Sigmoid()
tanh = nn.Tanh()
relu = nn.ReLU()
leaky_relu = nn.LeakyReLU()

ex1_tensor = torch.tensor([-1.0, 1.0, 0.0])

a1 = sigmoid(ex1_tensor)    # [0 , 1] 收斂在這個區間，容易梯度消失
a2 = tanh(ex1_tensor)       # [-1, 1] 相對於 sigmoid 均值是 0，會有更好的收斂效果
a3 = relu(ex1_tensor)       # [0 , x] relu 效果好但容易 die，因為小於 0 的都是 0，會梯度消失
a4 = leaky_relu(ex1_tensor) # [-n, x]  -n 非常小，防止梯度消失

print('Tensor \t\t', ex1_tensor, '\n')
print('Sigmoid \t', a1)
print('Tanh \t\t', a2)
print('ReLU \t\t', a3)
print('LReLU \t\t', a4)
```

# NN basic

```python
%matplotlib inline
import torch
import matplotlib.pyplot as plt
from torch import nn
import numpy as np

""" 建立一個複雜一點的數據 """
n, d = 200, 1
x = torch.rand(n,d)  # random 出
y = torch.sin(np.pi * x) * torch.cos(3*np.pi*x**2) # 正解的方程式，輸出是 y

plt.scatter(x.numpy(), y.numpy())  # 表格的尺規
plt.title('plot of $f(x)$')        # $數學式$
plt.xlabel('$x$'), plt.ylabel('$y$')
plt.show()
```

```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from torch import optim
import numpy as np
from matplotlib import animation

""" 建置一個神經網路 """
class Model(nn.Module):
    
    def __init__(self, d_in, d_out, act):    # 初始化的時候給予 d_in、d_out、act三個參數
        super().__init__()                   # 繼承 nn.Module 的 init
        self.l1 = nn.Linear(d_in, 64)        # 輸入維度為 [200,1] 所以第一層維度為 [1, n] 
        self.l2 = nn.Linear(64, 64)          # n 可以想像是神經元的數量
        self.l3 = nn.Linear(64, d_out)       # 最後的輸出層維度會根據數據的不同而改變，這邊維度為 1
        
        # 客製化選取激勵函數的部分 XD
        if(act=='relu'):
            print('Activation Fn : ReLU')
            self.act = nn.ReLU()
        elif(act=='tanh'):
            print('Activation Fn : Tanh')
            self.act = nn.Tanh()
        else:
            print('Activation Fn : None')
        
    def forward(self, x):
        
        if(act=='none'):                      # 如果沒有激勵函數
            out = self.l1(x)                  # 第一層輸入為 x
            out = self.l2(out)                # 第二層輸入為上一層的 out
            out = self.l3(out)                # 以此類推
            return out                        # 最後返回計算完的結果 
        
        else:                                 # 如果有激勵函數
            out = self.l1(x)
            out = self.act(out)               # 在傳遞給下一層之前要經過激勵函數 
            out = self.l2(out)
            out = self.act(out)
            out = self.l3(out)
            out = self.act(out)
            return out

""" 宣告神經網路以及參數 """
n_epochs = 6000                                 # 迭代次數
d_in, d_out = 1, 1                              # 輸入輸出維度
lr = 0.05                                       # 學習率 
t_log = n_epochs / 20                           # 總共紀錄 20 次

ls_act = ['relu', 'tanh', 'none']               # 激勵函數的清單
act = ls_act[2]                                 # 選擇激勵函數

model = Model(d_in, d_out, act )                # 實例化神經網路模型
loss_fn = nn.MSELoss()                          # 實例化損失函數
opt = optim.SGD(model.parameters(), lr = lr)    # 宣告優化器

""" 畫圖的部分 """
x_grid = torch.unsqueeze(torch.from_numpy(np.linspace(0,1,100)), 1).float()
fig = plt.figure(dpi=100)
plt.scatter(x.numpy(), y.numpy())  # 表格的尺規
plt.title('plot of $f(x)$')        # $數學式$
plt.xlabel('$x$'), plt.ylabel('$y$')
imgs=[]

""" 訓練 """
for i in range(n_epochs + 1):
    
    y_ = model(x)
    loss = loss_fn(y, y_)
    loss.backward()
    opt.step()
    opt.zero_grad()
    
    if(i % t_log==0):
        
        print('[{}/{}] loss:{:.4}'.format(i, n_epochs, loss.detach().numpy()))
       
        line = plt.plot(x_grid.detach().numpy(), model(x_grid).detach().numpy() , 'r')    # 畫出曲線
        imgs.append(line)

""" 儲存GIF """
ani = animation.ArtistAnimation(fig, imgs, interval=100, blit=True,
                                repeat_delay=2000)    
    
ani.save("non_linear_regression_{}_{}.gif".format('SGD', act ) , writer='pillow')
```
